```r
d <- data.rubis(where="duration = 60 and name like 'v0.29%' and nthreads <= 96 and total_time < 65")
```

```r
s <- sql("SELECT nthreads, cc_ph, mix, avg(rubis_UserSummary_count) as User, avg(rubis_NewBid_count) as Bid, avg(rubis_AddUser_count) as AddUser, avg(rubis_ViewAuction_count) as View, avg(rubis_CloseAuction_count) as Close, avg(rubis_BrowseItems_count) as Browse, avg(rubis_OpenAuction_count) as Open, avg(rubis_NewComment_count) as Comment FROM d GROUP BY nthreads, cc_ph, mix")
```

```r
s$total <- s$User+s$Bid+s$AddUser+s$View+s$Close+s$Browse+s$Open+s$Comment
```

```r
sql("SELECT (User+AddUser)/total, Bid/total, (View+Browse)/total, (Open+Close)/total, Comment/total FROM s WHERE nthreads = 64 ORDER BY mix")
```

# Retwis

```r
d <- data.retwis(where="(name like '%v0.30%' and nclients = 4)")
d <- subset(d, cc_ph %in% c(RW+PH, COMM, COMM+PH, COMB+PH, NOTXN))
d$cc_ph <- factor(d$cc_ph, levels = rev(c(RW+PH, COMM, COMM+PH, COMB+PH, NOTXN)))

d$x <- d$nthreads * num(d$nclients)
d$label <- d$nthreads * num(d$nclients) + "x" + d$rate
d$facet <- with(d, workload) # + "\n" + timeout_scaling)
```

```r
d <- subset(d, async == 0 & txn_failed < 200 & total_time > 60 & total_time < 65 & (phasing == 'on' | disable_txns == 1))
```

```r
sql("SELECT nthreads, rate, mix, alpha, throughput FROM d")
```

```r
s <- sql("SELECT nthreads, workload, cc_ph, avg(retwis_repost_count) as repost, avg(retwis_newuser_count) as newuser, avg(retwis_follow_count) as follow, avg(retwis_timeline_count) as timeline,avg(retwis_post_count) as post, avg(retwis_txn_count) as txn FROM d GROUP BY nthreads, workload, cc_ph")
```

```r
s$total <- s$repost+s$post+s$newuser+s$follow+s$timeline
```

```r
sql("SELECT workload, newuser/total, follow/total, post/total, repost/total, timeline/total FROM s WHERE nthreads = 64")
```
